```python
import pandas as pd
import numpy as np
df = pd.read_csv('../clean/Immigraton Data - Akshini Nithi.csv', encoding='utf-8-sig')
df.head()
df.shape
df.columns
df.info() 
df.describe()
df.duplicated()
```

```python
df_clean = df.drop_duplicates()
df_selected = df_clean[[
    'REF_DATE', 'GEO', 'DGUID',
    'Age (15C)', 'Gender (3)', 'Statistics (2)',
    'Immigrant status and period of immigration (9):Total - Immigrant status and period of immigration[1]',
    'Immigrant status and period of immigration (9):Non-immigrants[2]',
    'Immigrant status and period of immigration (9):Immigrants[3]',
    'Immigrant status and period of immigration (9):2016 to 2021[8]',
    'Immigrant status and period of immigration (9):Non-permanent residents[9]'
]]
df_selected = df_selected.rename(columns={
    'Age (15C)': 'Age',
    'Gender (3)': 'Gender',
    'Statistics (2)': 'Statistics',
    'Immigrant status and period of immigration (9):Total - Immigrant status and period of immigration[1]': 'Total',
    'Immigrant status and period of immigration (9):Non-immigrants[2]': 'Non_Immigrants',
    'Immigrant status and period of immigration (9):Immigrants[3]': 'Immigrants',
    'Immigrant status and period of immigration (9):2016 to 2021[8]': 'Recent_Immigrants_2016_2021',
    'Immigrant status and period of immigration (9):Non-permanent residents[9]': 'Non_Permanent_Residents'
})
df_selected = df_selected[df_selected['REF_DATE'] == 2021]
df_selected = df_selected[df_selected['Statistics'] == '2021 Counts']
df_selected.to_csv('../clean/Immigration - Akshini Nithi.csv', index=False)
```
