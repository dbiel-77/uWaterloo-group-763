```python
filtered_columns = ['REF_DATE', 'GEO', 'DGUID', 'Age (15C)', 'Gender (3)', 'Statistics (2)',
        'Visible minority (15):Total - Visible minority[1]',
        'Visible minority (15):Total visible minority population[2]',
       'Visible minority (15):South Asian[3]',
       'Visible minority (15):Chinese[4]',
       'Visible minority (15):Black[5]',
       'Visible minority (15):Filipino[6]',
       'Visible minority (15):Arab[7]',
       'Visible minority (15):Latin American[8]',
       'Visible minority (15):Southeast Asian[9]',
       'Visible minority (15):West Asian[10]',
       'Visible minority (15):Korean[11]',
       'Visible minority (15):Japanese[12]',
       'Visible minority (15):Visible minority, n.i.e.[13]',
       'Visible minority (15):Multiple visible minorities[14]',
       'Visible minority (15):Not a visible minority[15]']

filtered_scenarios = ['% distribution (2021)']
filtered_scenarios1 = ['Total - Gender']
filtered_scenarios2 = ['Total - Age']

df_filtered = df[filtered_columns]

df_filtered = df_filtered[df_filtered["Statistics (2)"].isin(filtered_scenarios)]
df_filtered = df_filtered[df_filtered["Gender (3)"].isin(filtered_scenarios1)]
df_filtered = df_filtered[df_filtered["Age (15C)"].isin(filtered_scenarios2)]
```
